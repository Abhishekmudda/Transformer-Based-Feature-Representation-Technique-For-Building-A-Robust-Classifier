Correlation Analysis for Numerical Features and Chi-Square Test for Categorical Features

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import chi2
from scipy.stats import pearsonr


numerical_features = X_res.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_features = X_res.select_dtypes(include=['object', 'category']).columns.tolist()
correlation_threshold = 0.1
correlated_features = []

for feature in numerical_features:
    corr, _ = pearsonr(X_res[feature], y_res)
    if abs(corr) >= correlation_threshold:
        correlated_features.append((feature, abs(corr)))

correlated_features_sorted = sorted(correlated_features, key=lambda x: x[1], reverse=True)
correlated_features_sorted = [feature for feature, _ in correlated_features_sorted]

print(f"Numerical features with strong correlation (sorted by correlation): {correlated_features_sorted}")

chi2_selected_features = []

for feature in categorical_features:
    chi2_score, p_value = chi2(X_res[feature].values.reshape(-1, 1), y_res)

    if p_value < 0.05:
        chi2_selected_features.append((feature, p_value))

chi2_selected_features_sorted = sorted(chi2_selected_features, key=lambda x: x[1])
chi2_selected_features_sorted = [feature for feature, _ in chi2_selected_features_sorted]

print(f"Categorical features selected by chi-square test (sorted by p-value): {chi2_selected_features_sorted}")
```

**Creating Embeddings**

```python
import torch
import torch.nn as nn
import pandas as pd

embedding_dim = 32
embedding_class = nn.Embedding

top_20_features = [
    'sttl', 'ct_state_ttl', 'Ltime', 'Stime', 'Dload', 'dttl', 'tcprtt',
    'dmeansz', 'ackdat', 'synack', 'dsport', 'Dpkts', 'Sload', 'ct_dst_ltm',
    'dbytes', 'state', 'Dintpkt', 'ct_src_ ltm', 'ct_dst_sport_ltm', 'ct_src_dport_ltm'
]

X_top_20 = X_res[top_20_features]
print(f"Shape of training dataset with top 20 features: {X_top_20.shape}")

feature_mins = {}
feature_maxs = {}
input_embeddings = []
embedding_layers = {}

for column in X_top_20.columns:
    feature_values = X_top_20[column].values
    feature_mins[column] = feature_values.min()
    feature_maxs[column] = feature_values.max()
    min_value = feature_values.min()
    feature_indices = feature_values - min_value

    unique_values = len(set(feature_values))
    max_index = feature_indices.max()
    num_embeddings = max(max_index + 1, unique_values)

    if column not in embedding_layers:
        embedding_layers[column] = embedding_class(num_embeddings=num_embeddings, embedding_dim=embedding_dim)

    feature_indices_tensor = torch.tensor(feature_indices, dtype=torch.long)
    feature_embedded = embedding_layers[column](feature_indices_tensor)

    input_embeddings.append(feature_embedded)

final_input_embeddings = torch.cat(input_embeddings, dim=1)
print("Final input embeddings shape for training data:", final_input_embeddings.shape)
output_tensor = torch.tensor(y_res.values, dtype=torch.long)
output_unique_values = len(torch.unique(output_tensor))
output_embedding_layer = nn.Embedding(num_embeddings=output_unique_values, embedding_dim=embedding_dim)
output_embeddings = output_embedding_layer(output_tensor)
print("Output embeddings shape for training data:", output_embeddings.shape)

#--------------------------------------------------------------------------------------
# Save everything needed for testing
torch.save(embedding_layers, "/content/drive/MyDrive/major project/feature_embeddings")
torch.save(output_embedding_layer.state_dict(), "/content/drive/MyDrive/major project/output_embedding")
torch.save({
    'mins': feature_mins,
    'maxs': feature_maxs
}, "/content/drive/MyDrive/major project/feature_scaling_factors")

print("Embeddings and scaling factors saved successfully.")
torch.save(final_input_embeddings, "/content/drive/MyDrive/major project/train_final_embeddings")
torch.save(output_embeddings, "/content/drive/MyDrive/major project/train_output_embeddings")
#-----------------------------------------------------------------------------------------------------------------------#
# Now, for the test dataset (X_test and y_test)

X_test_top_20 = X_test[top_20_features]
print(f"Shape of test dataset with top 20 features: {X_test_top_20.shape}")
test_input_embeddings = []

for column in X_test_top_20.columns:
    feature_values = X_test_top_20[column].values
    min_value = feature_values.min()
    feature_indices = feature_values - min_value
    embedding_layer = embedding_layers[column]
    feature_indices_tensor = torch.tensor(feature_indices, dtype=torch.long)
    feature_embedded = embedding_layer(feature_indices_tensor)
    test_input_embeddings.append(feature_embedded)

final_test_input_embeddings = torch.cat(test_input_embeddings, dim=1)
print("Final input embeddings shape for test data:", final_test_input_embeddings.shape)

test_output_tensor = torch.tensor(y_test.values, dtype=torch.long)

test_output_embeddings = output_embedding_layer(test_output_tensor)
print("Output embeddings shape for test data:", test_output_embeddings.shape)

```

**Training Pre-trained DistilBERT Model**

```python
import torch
import torch.nn as nn
from transformers import DistilBertModel, DistilBertConfig
from torch.utils.data import DataLoader, TensorDataset, random_split

# Device configuration
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

class TabularDistilBERT(nn.Module):
    def __init__(self, input_dim, output_dim, embedding_dim=768, seq_length=20):
        super(TabularDistilBERT, self).__init__()
        self.seq_length = seq_length
        self.embedding_dim = embedding_dim
        self.features_per_position = input_dim // seq_length

        # DistilBERT configuration
        self.distilbert = DistilBertModel.from_pretrained("distilbert-base-uncased")
        self.input_projection = nn.Linear(self.features_per_position, embedding_dim)
        self.positional_embeddings = nn.Embedding(seq_length, embedding_dim)
        self.fc = nn.Linear(embedding_dim, output_dim)
        self.dropout = nn.Dropout(0.1)

    def forward(self, x):
        batch_size = x.size(0)
        x = x.view(batch_size, self.seq_length, -1)
        x_projected = self.input_projection(x)
        position_ids = torch.arange(self.seq_length, device=x.device).unsqueeze(0).expand(batch_size, -1)
        pos_embeds = self.positional_embeddings(position_ids)
        x_with_pos = x_projected + pos_embeds
        bert_output = self.distilbert(inputs_embeds=x_with_pos)
        cls_output = bert_output.last_hidden_state[:, 0, :]
        cls_output = self.dropout(cls_output)
        logits = self.fc(cls_output)
        return logits

# Data preparation
train_X = torch.tensor(final_input_embeddings, dtype=torch.float32).to(device)
train_y = torch.tensor(output_tensor, dtype=torch.long).to(device)

dataset = TensorDataset(train_X, train_y)
train_size = int(0.8 * len(dataset))
val_size = len(dataset) - train_size
train_dataset, val_dataset = random_split(dataset, [train_size, val_size])

batch_size = 32
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size)

# Initialize model
input_dim = final_input_embeddings.shape[1]
model = TabularDistilBERT(input_dim=input_dim, output_dim=2).to(device)

# Loss and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.AdamW(model.parameters(), lr=2e-5)

# Training loop
num_epochs = 5
for epoch in range(num_epochs):
    model.train()
    train_loss, train_correct, train_total = 0, 0, 0

    for batch_X, batch_y in train_loader:
        optimizer.zero_grad()
        outputs = model(batch_X)
        loss = criterion(outputs, batch_y)
        loss.backward()
        optimizer.step()

        _, predicted = torch.max(outputs.data, 1)
        train_total += batch_y.size(0)
        train_correct += (predicted == batch_y).sum().item()
        train_loss += loss.item()

    # Validation
    model.eval()
    val_loss, val_correct, val_total = 0, 0, 0

    with torch.no_grad():
        for batch_X, batch_y in val_loader:
            outputs = model(batch_X)
            loss = criterion(outputs, batch_y)

            _, predicted = torch.max(outputs.data, 1)
            val_total += batch_y.size(0)
            val_correct += (predicted == batch_y).sum().item()
            val_loss += loss.item()

    print(f'Epoch [{epoch+1}/{num_epochs}]')
    print(f'Train Loss: {train_loss/len(train_loader):.4f}, Train Acc: {100*train_correct/train_total:.2f}%')
    print(f'Val Loss: {val_loss/len(val_loader):.4f}, Val Acc: {100*val_correct/val_total:.2f}%')
```
